## variable (var, let, const)


`var` -  ecmascript এর প্রথমদিকে ব্যাবহার করা হতো । var হচ্ছে global variable অর্থাৎ if-else statement এর ভেতর কোন variable declare করা হলে এর বাহিরেও যদি একি নামে কোন variable declare করা হয় তখন সেটা error হবে । এখন var সাধারণত ব্যাবহার করা হয় না ।  

`let, const` - 

1. let এবং const block level variable. let ভেরিয়েবলের মান সাধারণত চেঞ্জ করা যায়(অর্থাৎ reassign value) কিন্তু const এর মান চেঞ্জ করা যায় না ।   

2. let দিয়ে ভেরিয়েবল ডিক্লেয়ার করলে শুধু ডিক্লেয়ার করলেই হয়ে যায় কিন্তু initialize এর প্রয়োজন পরে না কিন্তু const দিয়ে ডিক্লেয়ার করলে অবশ্যই initialize করে দিতে হবে নাহলে error দিবে । ex :  

```javascript
// let
let score
score = 10
console.log(score)

// output: 10

// const
const score
score = 10
console.log(score)

// output : error
// correct : const score = 10;
```

## Arrays

```javascript
// declaration type-1 :  
let values = new Array(1,2,3,4,5)
console.log(values);


// decleration type-2 :
const values = [1,2, 'jahangir',4.5, [3.4, 'alam'], true, null]
console.log(values);

const fruits = ['jackfruits', 'apple', 'orange']
console.log(fruits)

// access values
console.log(fruits[1])

// add values
fruits[3] = 'bananna'
console.log(fruits)
// fruits = [];  error দিবে 


// push()
fruits.push('mango') // push value in the last

// unshift()
fruits.unshift('strawberries') // add value at the beginning


// pop()
fruits.pop(); // delete the last value


// check Array or not
console.log(Array.isArray(fruits))
console.log(Array.isArray('hello'))


// get index
console.log(fruits.indexOf('oranges')) // return 2



```

## for, while, for-of, forEach, map, filter loop

### for loop

```javascript
for(let i=0; i<=10; i++){
    console.log(`for loop number : ${i}`); // string literals
}  
// or

for(let i=0; i< todos.length; i++){
    console.log(todo[i].text);
}
```  

#### while loop  

```javascript

let i = 0;
while(i < 10){
    console.log(`While loop number : ${i}`);
}
```

#### for-of loop  

```javascript
for(let todo of todos){
    console.log(todo.text);
    console.log(todo.id);
}

```

#### forEach

```javascript

todos.forEach(function(todo){
    console.log(todo.text);
})
```

#### map

map returns an array

```javascript
const todoText = todos.map(function(todo){
    return todo.text;
})

console.log(todoText)

```

#### filter

```javascript
const todoCompleted = todo.filter(function(todo){
    return todo.isCompleted === true
}).map(function(todo){
    return todo.text
})


console.log(todoCompleted);
```

**output : যেগুলো অবজেক্ট true শুধু সেগুলোই নিবে এবং শুধু টেক্সট অংশটাই নিবে ।**


## Switches

```javascript
const x = 10;
const color = x > 12 ? 'red' : 'blue'
console.log(color)

switch(color){
    case 'red':
        console.log('color is red')
        break
    case 'blue':
        console.log('color is blue');
        break
    default:
        console.log('color is NOT red or blue');
        break
}
```
