# map() :

map() হচ্ছে Arrayর মেথড, এটির মেইন কাজ হলো অ্যাপ্লাইকৃত অ্যারের প্রত্যেকটা আইটেমের উপর একটা ফাংশন অ্যাপ্লাই করা।

আপনি map() কোনো অ্যারেতে কল করলে সেটা সেই অ্যারের প্রত্যেকটা আইটেমের উপর আপনারই দেওয়া একটা ফাংশন অ্যাপ্লাই করবে, এবং সবশেষে সবগুলো আবার আরেকটা অ্যারে হিসেবে রিটার্ন করবে।

```javascript
arr = [1,2,3,4]

function getSquare(item) {
   return item * item
}

arr.map(getSquare)

//output: 1,4,9,16

//new syntax
const doubleVal = newArray.map(item => item*2)
```

# filter():

ফিল্টার সেইম ম্যাপের মতোই কিন্তু এখানে অ্যাপ্লাইকৃত ফাংশন যদি `true` রিটার্ণ করে তাহলেই সেটা নতুন অ্যারেতে স্থান পাবে, নতুবা যদি ফাংশন `false` রিটার্ণ করে তাহলে সেই আইটেম অ্যারেতে স্থান পাবে না।

ধরুন আমি নিচের অ্যারে থেকে শুধুমাত্র জোড় নাম্বার গুলো বের করে নতুন অ্যারেতে রাখতে চাচ্ছিঃ

```javascript

var arr = [1, 2, 3, 4, 5, 6, 7, 8, 9]
const newArr = arr.filter(item => item % 2 === 0);

// output : 2,4,6,8
```

# reduce():

রিডিউস ও ম্যাপ, ফিল্টার এর মতো হলেও এটা একটা আর্গুমেন্ট বেশী নেয়। আমরা প্রোগ্রামিং ওয়ার্ল্ড এ স্ট্যাট ভ্যারিয়েবলের কথা জানি। এর কাজ হচ্ছে কোনো কিছুর কাউন্ট করা, বা ডিটেক্ট করে রাখা কোনো অপারেশনের চলার সময়। সহজ কথায় বললে আমরা যদি কোনো অ্যারে থেকে অ্যারের সব নাম্বারের টোটাল বের চাই, তাহলে সেক্ষেত্রে রিডিউস ইউজ করে সহজেই করে ফেলতে পারি। 

```javascript

var arr = [1, 2, 3, 4]
const t = arr.reduce((sum, item) => sum += item, 0)

// output : 10
```


# ড্রাই প্রিন্সিপ্যাল : 

প্রোগ্রামিং এ অনেক প্রিন্সিপ্যাল আছে। তার মধ্যে অন্যতম একটা হচ্ছে ড্রাই প্রিন্সিপ্যাল । ড্রাই প্রিন্সিপ্যাল অথবা ইংরেজীতে DRY মানে হচ্ছে Don’t Repeat Yourself. এর আসল উদ্দেশ্য হচ্ছে বারবার একই কাজ না করে সেটা কিভাবে শর্ট করে, সুন্দর করে করা যায়।  

ধরি আমরা একটা লেখা প্রিন্ট করতে চাচ্ছি। এখন বারবার সেটা না লিখে একটা ফাংশন ডিক্লেয়ার করে সেটা বারবার ইউজ করতে পারিঃ

```javascript

function addTwoNum(x, y) {
   var sum = x + y;
   console.log('Sum of ' + x + ' and ' + y + ' is: ' + sum);
}


addTwoNum(5, 10);
addTwoNum(20, 30);
addTwoNum(40, 50);
addTwoNum(60, 70);
addTwoNum(80, 90);

// output : 15, 50, 90,130, 170
```

# ক্লোজারস(Closures) নিয়ে ধারণা

আমরা জানি জাভাস্ক্রিপ্ট এ আপনি চাইলে ফাংশন থেকেও আরেকটা ফাংশন রিটার্ন করতে পারবেন। আর তখনি তৈরী হয় প্যারেন্ট-চাইল্ড রিলেশন। আর রিটার্নকৃত ফাংশনের সবকিছু চাইল্ড ফাংশন চাইলেও অ্যাক্সেস করতে পারে ক্লোজারস এর কারণে।

```javascript
function closureDemo(){
    let x = 10
    return function(){
        let y = 20
        console.log('sum : '+ (x+y));
        
    }
}

closureDemo()()

/*

this can call with different way

var aVar = closuresDemo();
var total = aVar();
console.log(total);

*/


function aParentFunc(a) {
   return function(b) {
      console.log('Sum: ' + (a + b));
   }
}

aParentFunc(6)(4);
```

# Immediately Invoked Function Expressions (IIFE) 

আমাদের সাধারনত কোনো ফাংশন বানানোর পর পরে সেটাকে কল করে ইউজ করতে হয়। কিন্তু যদি আমরা ফাংশন বানানোর সাথে সাথেই সেটাকে কল করতে চাই সেক্ষেত্রে আমরা ইফি Immediately Invoked Function Expressions (IIFE) টেকনিক ইউজ করতে পারি। 

ইফি তে সাধারণত পুরো ফাংশনটাকে প্রথম ব্র্যাকেটস এর ভিতরে রাখতে হয় এবং সবশেষে আরো দুইটা আর্গুমেন্ট ব্র্যাকেটস দিয়ে কল করতে হয়। উদাহরন দেখলে ক্লিয়ার হয়ে যাবে। ধরি কোন ফাংশন আমি সরাসরি ক্রিয়েট করে সাথে সাথেই কল করতে চাইঃ

```javascript
(function aDemoFunc() {  
   console.log('Hello World!');  
})();
```
এখানে পুরো ফাংশনটা প্রথম ব্র্যাকেটস এর ভিতরে চলে যাবে। এবং সবশেষে আরো দুইটা ব্র্যাকেটস হবে যেখানে যদি উক্ত ফাংশনের কোনো আর্গুমেন্ট থাকে তাহলে পাস করতে পারবেন। 

expression এও এটি কাজ করবে 

```javascript
var sum = (function() {
   return 10 + 20; 
})();

```

### পার্থক্য নং - ১ 
`var` কীওয়ার্ড ফাংশন স্কোপ মেইন্টেইন করে  
`let` ও `const` ব্লক(Block) স্কোপ মেইন্টেইন করে  

এর মানে হচ্ছে কোন ফাংশনের বাহিরে `var` দিয়ে ডিক্লেয়ার করা ভেরিয়েবলের এক্সেস নাই অন্যদিকে `let` এবং `const` দিয়ে ডিক্লেয়ার করা মানে হচ্ছে যেকোনো স্কোপের বাহিরে ওই ভেরিয়েবলের এক্সেস নাই, সেই স্কোপ হতে পারে কোন if block, switch or for loop block । 

```javascript

function es5var() {
   if(true) {
      var a = 5;
   }
   console.log('Value: ' + a);
}

es5var();

```
এখানে `var` দিয়ে ডিক্লেয়ার করার কারণে ফাংশনের মধ্যে হওয়াতে if ব্লকের বাহিরে এক্সেস পাবে কিন্তু যদি `let` or `const` ডিক্লেয়ার করলে if ব্লকের বাহিরে আর এক্সেস পাইতো না ।
