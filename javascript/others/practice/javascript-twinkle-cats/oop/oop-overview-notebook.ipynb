# this keyowrd 

```javascript

var obj = {
  value : 20
}

function add(num){
  console.log(this.value);
  
  return this.value + num;
}

let binded = add.bind(obj)
let result = binded(100)

console.log(result);

```

যখন method এর ভেতর this keyword use করা হবে তখন সরাসরি স্কোপের অবজেক্ট কে রেফার করবে  কিন্তু function এর ভেতর this keyword টা উইন্ডো বা গ্লোবাল অবজেক্টকে রেফার করবে । কারণ this সরাসরি object এর সাথে সম্পর্কযুক্ত । 



### prototype

```javascript

function Person(name, age){
  this.name = name;
  this.age = age;
}

Person.prototype.hello = function(){
  console.log('Hello, ' + this.name );
}

let p1 = new Person('jahangir', 26);
let p2 = new Person('imran', 25);

console.log(p1);
console.log(p2);

console.log(Person.prototype);

```

দুইটা অবজেক্ট একি মেথডকে অ্যাক্সেস করে কিনা এটা চেক করার জন্য `p1.__proto__ == p2.__proto__` এই কন্ডিশন যদি true return করে তাহলে একি মেথড অ্যাক্সেস করে 

prototype হচ্ছে একটা অবজেক্ট ।  

![prototype](./images/proto-2.png)  

prototype এ একটা ফাংশন বা ভেরিয়েবল ডিক্লেয়ার হচ্ছে কিন্তু যে কোন অবজেক্ট সেটা শেয়ার করছে অর্থাৎ অবজেক্টগুলোর মধ্যে কমন কোন অংশ থাকলে সেগুলো prototype এ ডিক্লেয়ার করা হয় । তাহলে কম মেমরি লস হয় । 

### bind(), call(), apply()
__bind()__

```javascript

function printme(){
  console.log('Hello, ' + this.name );
}

let obj1 = {
  name : 'Jahangir alam',
  age : 27
}


let obj2 = {
  name : 'Abdullah imran',
  email : 'imran@gmail.com'
}

let binded = printme.bind(obj2);

binded()

```
bind() function এর সাথে যে অবজেক্টটা দেওয়া হয় সেটা একসাথে হয়ে নতুন একটা ফাংশন রিটার্ন করে । সাধারণত কোন ফাংশনের ভেতরে যে this keyword টা দেওয়া হয় সেটা হচ্ছে গ্লোবাল অর্থাৎ window. কিন্তু যখন bind function এর মাধ্যমে জয়েন করা হবে তখন সেই অবজেক্টের এলিমেন্ত গুলকে রেফার করবে this keyword এর মাধ্যমে ।   

__call()__  
bind() function এ একটা সমস্যা ছিল যে আলাদা করে একটা ভেরিয়েবল নিতে হতো , কারণ bind() function আরেকটা ফাংশন রিটার্ন করে তাই আলাদাভাবে অন্ন ভেরিয়েবল এ ভেলুটা রেখে ওই ভেরিয়েবলটা প্রিন্ট করতে হয় । কিন্তু call() function কল করলে আলাদাভাবে ভেরিয়েবল নিতে হয় না ।  

`printme.call(obj1)`  

__apply()__   

```javascript

function add(a,b){
  return (a+b) * this.c;
}

let obj1 = {
  c : 3
}

let obj2 = {
  c : 5
}

// call()
console.log(add.call(obj1, 10,4));

// apply()
console.log(add.apply(obj1, [10,4]));

// bind()
let binded = add.bind(obj1)
console.log(binded(10,4));

```

এখানে `Student.prototype = Object.create(Person.prototype);` এই লাইনটা ইউজ করার ফলে Person এর constructor টা Student এর constructor কে override করে ফেলে । অর্থাৎ তখন person er constructor set হয়ে যায় ।
